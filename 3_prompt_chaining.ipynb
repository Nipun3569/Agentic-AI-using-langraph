```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
from langchain_ollama import ChatOllama
```

```python
model=ChatOllama(model="gemma:2b")
```

```python
class blogState(TypedDict):
    topic:str
    outline:str
    blog:str
```

```python
def create_outline(state:blogState)->blogState:
    #fetch title
    title=state['topic']
    #call llm to gen outline
    prompt=f"Create a detailed outline for a blog post with the title: {title}"
    outline=model.invoke(prompt)
    #pdate state
    state['outline']=outline
    return state
```

```python
def write_blog(state:blogState)->blogState:
    title=state['topic']
    outline=state['outline']

    prompt=f"Write a blog post with the title: {title} using the following outline: {outline}"

    content=model.invoke(prompt)
    state['blog']=content
    return state
```

```python
graph=StateGraph(blogState)

#nodes
graph.add_node('create_outline',create_outline)
graph.add_node('write_blog',write_blog)

#create edges
graph.add_edge(START,'create_outline')
graph.add_edge('create_outline','write_blog')
graph.add_edge('write_blog',END)

workflow= graph.compile()
```

```python
initial_state=blogState(topic="Rise of Ai in India")
final_state=workflow.invoke(initial_state)
print(final_state['outline'])
print(final_state['blog'])
```
